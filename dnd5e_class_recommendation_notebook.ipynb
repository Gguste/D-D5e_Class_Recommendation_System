```julia
function norma(vetor)
    tam = vetor'*vetor
    return sqrt(tam)
end

function distancia(vetor1, vetor2)
    return norma(vetor1-vetor2)
end

function cosseno(vetor1, vetor2)
    return (vetor1'*vetor2)/(norma(vetor1)*norma(vetor2))
end
```

```julia
function readCSV()
    df = CSV.read("D&D5e_dataset.csv", DataFrame;drop=[1]) #drop para não pegar a coluna da data/hora
    databaseText = Matrix(df)
    num_player, num_subclass = size(databaseText) #pega o número de linhas(players) e o número de colunas(subclasses)
    database = zeros(num_player,num_subclass)
    for i in 1:num_player
        for j in 1:num_subclass
            if(cmp(databaseText[i,j], "Never played / Nunca jogou") == 0)
                database[i,j] = 0.0 #será tratado como Neutral / Indiferente no momento
            end
            if(cmp(databaseText[i,j], "Detest / Odeia") == 0)
                database[i,j] = -2.0
            end
            if(cmp(databaseText[i,j], "Dislike / Não gosta") == 0)
                database[i,j] = -1.0
            end
            if(cmp(databaseText[i,j], "Neutral / Indiferente") == 0)
                database[i,j] = 0.0
            end
            if(cmp(databaseText[i,j], "Like / Gosta") == 0)
                database[i,j] = 1.0
            end
            if(cmp(databaseText[i,j], "Love / Adora") == 0)
                database[i,j] = 2.0
            end
        end
    end
    return database
end
database = readCSV()
```

```julia
function recommendationByPlayers_Cosseno(person,database) #Compara uma pessoa a todas e dps recomenda as classes favoritas da pessoa mais similar
    num_player, num_subclass = size(database)
    bestOption = 0
    bestScore = 0.0
    for i in 1:num_player
        score = cosseno(database[i,:],person)
        if ((score > bestScore) || (i == 1))
            bestOption = i
            bestScore = score
        end
    end
    println("Classes Recomendadas:")
    for i in 1:num_subclass
        if((person[i] == 0) && ((database[bestOption,i] > 0)))
            println(getFullName(subClassInfo[i]))
        end
    end
end

function recommendationByPlayers_Distancia(person, database) #Compara uma pessoa a todas e dps recomenda as classes favoritas da pessoa mais similar
    num_player, num_subclass = size(database)
    bestOption = 0
    bestScore = 0.0
    for i in 1:num_player
        score = distancia(database[i,:],person)
        if (((score!= 0) && (score < bestScore)) || (i == 1)) #Mesmo eles não sendo o mesmo vetor, ainda ocorre de pegar outro usuário que marcou apenas as 3 classes do teste
            bestOption = i
            bestScore = score
        end
    end
    println("Classes Recomendadas:")
    for i in 1:num_subclass
        if((person[i] == 0) && ((database[bestOption,i] > 0)))
            println(getFullName(subClassInfo[i]))
        end
    end
end
```

```julia
function recommendationByClass(subclassIndex, database) #Compara a classe escolhida com as outras para criar um ranking Top 3 para ser recomendado
    num_player, num_subclass = size(database)
    bestOption = [0 0 0]
    bestScore = [-1.0 -1.0 -1.0]
    for i in 1:num_subclass
        score = cosseno(database[:,i],database[:,subclassIndex])
        if (i != subclassIndex)
            if (bestScore[1] < score)
                bestOption[3] = bestOption[2]
                bestScore[3] = bestScore[2]
                bestOption[2] = bestOption[1]
                bestScore[2] = bestScore[1]
                bestOption[1] = i
                bestScore[1] = score
            elseif (bestScore[2] < score)
                bestOption[3] = bestOption[2]
                bestScore[3] = bestScore[2]
                bestOption[2] = i
                bestScore[2] = score
            elseif (bestScore[3] < score)
                bestOption[3] = i
                bestScore[3] = score
            end
        end
    end
    println("Classes Recomendadas:")
    for i in 1:length(bestOption)
        println(string(i," - ",getFullName(subClassInfo[bestOption[i]])))
    end
end
```

```julia
using Statistics
function recommendationByGroupOfClasses(subclassIndex, database) #Compara com um grupo de subclasses escolhidas com as outras para criar um ranking Top 3 para ser recomendado
    num_player, num_subclass = size(database)
    bestOption = [0 0 0]
    bestScore = [-1.0 -1.0 -1.0]
    for i in 1:num_subclass
        score = convert(Array{Float64,1}, zeros(length(subclassIndex)))
        media = 0
        for j in 1:length(subclassIndex)
            score[j] = cosseno(database[:,i],database[:,subclassIndex[j]])
        end
        media = mean(score)
        if !(i in subclassIndex)
            if (bestScore[1] < media)
                bestOption[3] = bestOption[2]
                bestScore[3] = bestScore[2]
                bestOption[2] = bestOption[1]
                bestScore[2] = bestScore[1]
                bestOption[1] = i
                bestScore[1] = media
            elseif (bestScore[2] < media)
                bestOption[3] = bestOption[2]
                bestScore[3] = bestScore[2]
                bestOption[2] = i
                bestScore[2] = media
            elseif (bestScore[3] < media)
                bestOption[3] = i
                bestScore[3] = media
            end
        end
    end
    println("Classes Recomendadas:")
    for i in 1:length(bestOption)
        println(string(i," - ",getFullName(subClassInfo[bestOption[i]])))
    end
end
```

```julia
subclassChosedArray = convert(Array{Int64,1},zeros(3))
subclassChosedArray[1] = get(subClassDictionary,"Barbarian Path of the Ancestral Guardian",1)
subclassChosedArray[2] = get(subClassDictionary,"Cleric Light Domain",1)
subclassChosedArray[3] = get(subClassDictionary,"Ranger Gloom Stalker Conclave",1)
#subclassChosedArray[1] = 16
#subclassChosedArray[2] = 18
#subclassChosedArray[3] = 17
println(getFullName(subClassInfo[subclassChosedArray[1]]))
println(getFullName(subClassInfo[subclassChosedArray[2]]))
println(getFullName(subClassInfo[subclassChosedArray[3]]))
recommendationByGroupOfClasses(subclassChosedArray,database)
```
